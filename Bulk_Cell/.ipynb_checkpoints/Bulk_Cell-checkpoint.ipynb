```python
import pandas as pd

file_path = "pancreas_refseq_rpkms_wholeislet.txt"

total_rows = 0

for chunk in pd.read_csv(file_path, sep="\t", chunksize=1000):
    total_rows += len(chunk)

print("Total rows:", total_rows)
```
