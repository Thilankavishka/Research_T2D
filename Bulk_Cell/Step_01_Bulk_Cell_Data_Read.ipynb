```python
import pandas as pd

chunk_iter = pd.read_csv(
    "pancreas_refseq_rpkms_wholeislet.txt",
    sep="\t",
    engine="python",
    chunksize=1000
)

total_genes = 0
num_samples = None

# First chunk
first_chunk = next(chunk_iter)
total_genes += first_chunk.shape[0]

# Debugging
print("Columns detected:")
print(first_chunk.columns.tolist())
print("Total columns:", len(first_chunk.columns))

# Count sample columns
# Here first column is GeneName index, second is RefSeqID,
# then 7 expression columns and 7 raw count columns
num_samples = (len(first_chunk.columns) - 1) 

# Remaining chunks
for chunk in chunk_iter:
    total_genes += chunk.shape[0]

print("Total Genes:", total_genes)
print("Total Samples:", num_samples)
```

```python
all_samples = first_chunk.columns[1:]  # ONLY skip '#samples'

disease_samples = [c for c in all_samples if "T2D" in c]
control_samples = [c for c in all_samples if "T2D" not in c]

print("Control samples:", len(control_samples))
print("Disease samples:", len(disease_samples))
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Genes vs Samples
ax1.bar(["Genes", "Samples"], [total_genes, num_samples], color=['#4C72B0', '#55A868'])
for i, v in enumerate([total_genes, num_samples]):
    ax1.text(i, v, str(v), ha='center')

ax1.set_title("Bulk RNA Dataset Composition")

# Control vs Disease
ax2.bar(["Control", "Disease(T2D)"], [len(control_samples), len(disease_samples)],
        color=['#4C72B0', '#C44E52'])
for i, v in enumerate([len(control_samples), len(disease_samples)]):
    ax2.text(i, v, str(v), ha='center')

ax2.set_title("Control vs Disease(T2D) Samples")

plt.tight_layout()
plt.savefig("bulkRNA_summary.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file
file_path = "E-MTAB-5060.sdrf.txt"
df = pd.read_csv(file_path, sep="\t")

# ----------------------------
# Basic info
# ----------------------------
print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())

# ----------------------------
# Count summaries
# ----------------------------
print("\nDisease counts:")
print(df["Factor Value[disease]"].value_counts())

print("\nSex counts:")
print(df["Characteristics[sex]"].value_counts())

print("\nIndividual counts:")
print(df["Characteristics[individual]"].value_counts())

# ----------------------------
# Colors
# ----------------------------
disease_colors = ["#F44336", "#4CAF50"]   # red / green
sex_colors = ["#2196F3", "#E91E63"]       # blue / pink
age_color = "#FF9800"
bmi_color = "#009688"

# ----------------------------
# Create subplots
# ----------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("E-MTAB-5060 Bulk RNA-seq Dataset Summary", fontsize=18)

# Disease count
df["Factor Value[disease]"].value_counts().plot(
    kind="bar", color=disease_colors, ax=axes[0,0]
)
axes[0,0].set_title("Sample Count by Disease")
axes[0,0].set_xlabel("Disease")
axes[0,0].set_ylabel("Count")
axes[0,0].tick_params(axis='x', rotation=20)

# Sex count
df["Characteristics[sex]"].value_counts().plot(
    kind="bar", color=sex_colors, ax=axes[0,1]
)
axes[0,1].set_title("Sample Count by Sex")
axes[0,1].set_xlabel("Sex")
axes[0,1].set_ylabel("Count")
axes[0,1].tick_params(axis='x', rotation=0)

# Age distribution
df["Characteristics[age]"].plot(
    kind="hist", bins=8, color=age_color, edgecolor="black", ax=axes[1,0]
)
axes[1,0].set_title("Age Distribution")
axes[1,0].set_xlabel("Age")

# BMI distribution
df["Characteristics[body mass index]"].plot(
    kind="hist", bins=8, color=bmi_color, edgecolor="black", ax=axes[1,1]
)
axes[1,1].set_title("BMI Distribution")
axes[1,1].set_xlabel("BMI")

plt.tight_layout(rect=[0,0,1,0.95])

# Save image
plt.savefig("bulk_dataset_summary.png", dpi=300, bbox_inches="tight")
plt.show()
```
