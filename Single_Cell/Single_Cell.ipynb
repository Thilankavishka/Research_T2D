```python
import pandas as pd

chunk_iter = pd.read_csv(
    "pancreas_refseq_rpkms_counts_3514sc.txt",
    sep="\t",
    chunksize=1000
)

total_genes = 0
num_cells = None

# Process first chunk
first_chunk = next(chunk_iter)
total_genes += first_chunk.shape[0]
num_cells = first_chunk.shape[1] - 1  # exclude '#samples' column

# Process remaining chunks
for chunk in chunk_iter:
    total_genes += chunk.shape[0]

print("Total Genes:", total_genes)
print("Total Cells:", num_cells)
```

```python
import matplotlib.pyplot as plt

# -----------------------------
# Create figure with 2 parts
# -----------------------------
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# -----------------------------
# Part 1: Genes vs Cells
# -----------------------------
colors1 = ['#4C72B0', '#55A868']  # professional blue & green

bars1 = ax1.bar(
    ["Genes (Rows)", "Cells (Columns)"],
    [total_genes, num_cells],
    color=colors1,
    edgecolor='black'
)

# Add values on top
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval,
             f'{int(yval)}', ha='center', va='bottom', fontsize=10)

ax1.set_ylabel("Count", fontsize=11)
ax1.set_title("Dataset Composition", fontsize=13, fontweight='bold')

# -----------------------------
# Part 2: Control vs Disease
# -----------------------------
colors2 = ['#4C72B0', '#C44E52']  # blue & red (control vs disease)

bars2 = ax2.bar(
    ["Control Cells", "Disease(T2D) Cells"],
    [len(control_cells), len(disease_cells)],
    color=colors2,
    edgecolor='black'
)

# Add values on top
for bar in bars2:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval,
             f'{int(yval)}', ha='center', va='bottom', fontsize=10)

ax2.set_ylabel("Number of Cells", fontsize=11)
ax2.set_title("Control vs Disease(T2D) Cells", fontsize=13, fontweight='bold')

# -----------------------------
# Final layout & save
# -----------------------------
plt.tight_layout()
plt.savefig("scRNA_Data_Summary.png", dpi=300)
plt.show()
```

```python
import pandas as pd

# File paths
input_file = "pancreas_refseq_rpkms_counts_3514sc.txt"
control_file = "scRNA_control_data.csv"
disease_file = "scRNA_disease_data.csv"

# Read first chunk to get column names
chunk_iter = pd.read_csv(input_file, sep="\t", chunksize=1000)
first_chunk = next(chunk_iter)

# Identify columns
all_cells = first_chunk.columns[1:]  # exclude '#samples'
disease_cells = [c for c in all_cells if "T2D" in c]
control_cells = [c for c in all_cells if "T2D" not in c]

# Create headers
control_columns = ['#samples'] + control_cells
disease_columns = ['#samples'] + disease_cells

# -----------------------------
# Write first chunk
# -----------------------------
first_chunk[control_columns].to_csv(control_file, index=False, mode='w')
first_chunk[disease_columns].to_csv(disease_file, index=False, mode='w')

# -----------------------------
# Process remaining chunks
# -----------------------------
for chunk in chunk_iter:
    chunk[control_columns].to_csv(control_file, index=False, mode='a', header=False)
    chunk[disease_columns].to_csv(disease_file, index=False, mode='a', header=False)

print("✅ Control data saved:", control_file)
print("✅ Disease data saved:", disease_file)
```

```python
import pandas as pd

# -----------------------------
# Control Data
# -----------------------------
control_chunks = pd.read_csv("scRNA_control_data.csv", chunksize=1000)

control_rows = 0
control_cols = None

for chunk in control_chunks:
    control_rows += chunk.shape[0]
    
    if control_cols is None:
        control_cols = chunk.shape[1] - 1  # exclude '#samples'

print("Control Data:")
print("Total Genes (rows):", control_rows)
print("Total Cells (columns):", control_cols)

# -----------------------------
# Disease Data
# -----------------------------
disease_chunks = pd.read_csv("scRNA_disease_data.csv", chunksize=1000)

disease_rows = 0
disease_cols = None

for chunk in disease_chunks:
    disease_rows += chunk.shape[0]
    
    if disease_cols is None:
        disease_cols = chunk.shape[1] - 1  # exclude '#samples'

print("\nDisease Data:")
print("Total Genes (rows):", disease_rows)
print("Total Cells (columns):", disease_cols)
```

```python
import matplotlib.pyplot as plt

labels = ["Control Cells", "Disease Cells"]
values = [control_cols, disease_cols]

plt.bar(labels, values)
for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center')

plt.title("Control vs Disease Cell Count")
plt.ylabel("Number of Cells")
plt.show()
```
