```python
# Install if needed
# pip install scanpy

import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
control_df = pd.read_csv("scRNA_control_data.csv", index_col=0)

# Ensure proper string indices
control_df.index = control_df.index.astype(str)

# Remove duplicate gene names
control_df = control_df[~control_df.index.duplicated()]

# Ensure column names (cells) are strings
control_df.columns = control_df.columns.astype(str)

# Convert to AnnData (Scanpy format)
adata = sc.AnnData(control_df.T)

adata.obs["condition"] = "control"

print(adata)
```

```python
#QC
```

```python
sc.pp.calculate_qc_metrics(adata, inplace=True)

# ==========================
# MANUAL QC METRICS
# ==========================

import numpy as np

# Total counts per cell
adata.obs["total_counts"] = np.sum(adata.X, axis=1)

# Number of genes detected per cell
adata.obs["n_genes_by_counts"] = np.sum(adata.X > 0, axis=1)
```

```python
import scanpy as sc

sc.pl.scatter(
    adata,
    x='total_counts',
    y='n_genes_by_counts',
    color='#6366F1',   # modern indigo
    alpha=0.7
)
```

```python
sc.pl.violin(
    adata,
    ['total_counts', 'n_genes_by_counts'],
    jitter=0.4,
    multi_panel=True,
    color="#3B82F6"   # single color only
)
```

```python
#STEP 1 - Filtering Cells and Genes
```

```python
before_shape = adata.shape
print("Before filtering:", before_shape)
```

```python
import scanpy as sc

sc.pl.violin(
    adata,
    ['total_counts', 'n_genes_by_counts'],
    jitter=0.4,
    multi_panel=True
)
```

```python
sc.pp.filter_cells(adata, min_genes=200)
```

```python
sc.pp.filter_genes(adata, min_cells=3)
```

```python
adata = adata[adata.obs['total_counts'] > 1000, :]
```

```python
sc.pl.violin(
    adata,
    ['total_counts', 'n_genes_by_counts'],
    jitter=0.4,
    multi_panel=True
)
```
